```python
import pandas as pd

df=pd.read_csv('sales_POI.csv')

from math import radians, cos, sin, asin, sqrt
 
def distance(lon1, lat1, lon2, lat2): 

    lon1, lat1, lon2, lat2 = map(radians, [lon1, lat1, lon2, lat2])
 
    dlon = lon2 - lon1 
    dlat = lat2 - lat1 
    a = sin(dlat/2)**2 + cos(lat1) * cos(lat2) * sin(dlon/2)**2
    c = 2 * asin(sqrt(a)) 
    r = 6371 # 地球平均半径，单位为公里
    return c * r * 1000
```

```python
material=pd.read_excel('material.xlsx')
```

```python
fake=list(material.iloc[0:23,0])
```

```python
brand_list=list(material.iloc[2:,1])
```

```python
del brand_list[0]
```

```python
fake_index=[]
for word in fake:
    for index_num, item in enumerate(list(df['FIX_NAME'])):
        if word in item:
            fake_index.append(index_num)
```

```python
fake=[]
for i in range(len(df)):
    if i in fake_index:
        fake.append('fake')
    else:
        fake.append(None)
```

```python
df['Fake']=fake
```

```python
a=[]
for i in range(len(df)):
    if type(df['BRAND_1'][i])==str:
        a.append(df['BRAND_1'][i])
    else:
        a.append(df['BRAND_2'][i])

df['BRAND']=a
```

```python
remove200=['世纪华联','好又多','华联_华联超市','兴盛','联合一百','快乐惠','联华','文峰','好邻居','锦和','华联','苏果_苏果便利','苏果','每天惠','东风','双汇','麦点','如海','美宜佳','国大','每天惠_智慧','易捷','佐客','中百','红旗','爱便利','中百_超市','天福','每一天','雅堂小超','科迪','迪亚天天','昆仑好客','唐久']
```

```python
remove500=[i for i in brand_list if i not in remove200]
```

```python
result=[]
index_result=[]
for brand in remove200:
    df_temp_sm=df.loc[(df['BRAND']==brand)&(df['SUBCATEGORY']=='超市')]

    ref_lon=df_temp_sm['DISPLAY_X']
    ref_lat=df_temp_sm['DISPLAY_Y']

    df_temp_other=df.loc[(df['BRAND']==brand)&(df['SUBCATEGORY']!='超市')]

    cal_lon=df_temp_other['DISPLAY_X']
    cal_lat=df_temp_other['DISPLAY_Y']

    sm_index=df_temp_sm.index
    other_index=df_temp_other.index

    
    
    for i in sm_index:
        for j in other_index:        
            dis=distance(ref_lon[i],ref_lat[i],cal_lon[j],cal_lat[j])
            if dis<=200:
                result.append(df_temp_other['﻿POI_ID'][j])
                index_result.append(j)
                print('重合距离：',dis,' 超市店面：',df['FIX_NAME'][i],' 非超市店面：',df['FIX_NAME'][j])
```

```python
for brand in remove500:
    df_temp_sm=df.loc[(df['BRAND']==brand)&(df['SUBCATEGORY']=='超市')]

    ref_lon=df_temp_sm['DISPLAY_X']
    ref_lat=df_temp_sm['DISPLAY_Y']

    df_temp_other=df.loc[(df['BRAND']==brand)&(df['SUBCATEGORY']!='超市')]

    cal_lon=df_temp_other['DISPLAY_X']
    cal_lat=df_temp_other['DISPLAY_Y']

    sm_index=df_temp_sm.index
    other_index=df_temp_other.index

    
    
    for i in sm_index:
        for j in other_index:        
            dis=distance(ref_lon[i],ref_lat[i],cal_lon[j],cal_lat[j])
            if dis<=500:
                result.append(df_temp_other['﻿POI_ID'][j])
                index_result.append(j)
                print('重合距离：',dis,' 超市店面：',df['FIX_NAME'][i],' 非超市店面：',df['FIX_NAME'][j])
```

```python
mark=[]
for i in range(len(df)):
    if i in index_result:
        mark.append('remove')
    else:
        mark.append(None)
```

```python
df['Mark']=mark
```
